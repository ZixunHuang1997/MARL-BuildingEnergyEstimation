```python
import os
import json
import ijson
import decimal
```

```python
class DecimalEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, decimal.Decimal):
            return float(obj)  # or str(obj) if you want it as a string
        return super(DecimalEncoder, self).default(obj)
```

```python
filename = "Countywide_Building_Outlines_2014.geojson"

# Define your size limit (in bytes)
size_limit = 50 * 1024 * 1024  # 50 MB

# Initialize variables
size_counter = 0
file_counter = 0
features = []

# Open your source GeoJSON file
with open(filename, 'r') as f:
    objects = ijson.items(f, 'features.item')
    
    # Iterate over each feature
    for obj in objects:
        # Convert the feature to a JSON string and get its length in bytes
        feature_str = json.dumps(obj, cls=DecimalEncoder)
        feature_size = len(feature_str.encode('utf-8'))

        # If adding this feature would exceed the size limit, write the current features to a file
        if size_counter + feature_size > size_limit:
            with open(f'geojson_{file_counter}.json', 'w') as out_file:
                # Create a GeoJSON structure with the features
                geojson_data = {
                    "type": "FeatureCollection",
                    "features": features
                }
                json.dump(geojson_data, out_file, cls=DecimalEncoder)
                print(f"Created file: geojson_{file_counter}.json")

            # Reset the features and size counter, and increment the file counter
            features = []
            size_counter = 0
            file_counter += 1

        # Add the feature to the list of features, and increment the size counter
        features.append(obj)
        size_counter += feature_size

# Don't forget to write out the last batch of features if there are any left
if features:
    with open(f'geojson_{file_counter}.json', 'w') as out_file:
        geojson_data = {
            "type": "FeatureCollection",
            "features": features
        }
        json.dump(geojson_data, out_file, cls=DecimalEncoder)
```
