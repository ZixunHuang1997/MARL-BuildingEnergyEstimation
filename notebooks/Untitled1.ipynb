```python
import pandas as pd
```

```python
import geopandas as gpd
```

```python
curr_data = 'zone_3_4'
zone03 = gpd.read_file(f'../data/data_root/{curr_data}.geojson')
utm_proj = "+proj=utm +zone=11 +datum=WGS84 +units=m +no_defs"
zone03 = zone03.to_crs(utm_proj)
zone03['area'] = zone03['geometry'].area
zone03 = zone03[['hgt_median_m', 'area', 'yrbuilt']]
zone03 = zone03.reset_index()
zone03.to_csv(f"../data/data_root/building_metadata_{curr_data}.csv", index=False)
zone03
```
