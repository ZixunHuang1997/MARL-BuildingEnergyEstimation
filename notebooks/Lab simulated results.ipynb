```python
def list_files_in_directory(directory):
    files = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            files.append(filename)
    return files

```

```python
# Specify the directory path
directory_path = "../data/data_idf/"

# Call the function to list files in the directory
file_list = list_files_in_directory(directory_path)
file_list[0]
```

```python
energy_df = pd.DataFrame(columns=['category', 'consumption'])

for file in tqdm(file_list):
    tmp_df = pd.read_csv(os.path.join(directory_path, file), index_col=None)
    elec_sum = tmp_df['Electricity:Facility [J](Hourly)'].sum()
    gas_sum = tmp_df['NaturalGas:Facility [J](Hourly) '].sum()
    energy_sum = elec_sum+gas_sum
    
    index_value = file.split('.')[0]
    energy_value = energy_sum
    energy_df.loc[len(energy_df)] = [index_value, energy_value]

energy_df
```

```python
energy_df.to_csv('energy_df.csv',index=False)
```

```python
energy_df[['Category1', 'epw']] = energy_df['category'].str.split('____', expand=True)
energy_df
```
